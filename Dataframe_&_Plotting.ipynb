### Setting up Python Environment

```python
import pandas as pd
```

###Indl√¶sning af en DataFrame

```python
#Importing the dataset
CSVDataFile="https://raw.githubusercontent.com/jpandersen61/Machine-Learning/refs/heads/main/50_Startups.csv"
StartUpsDataFrame = pd.read_csv(CSVDataFile)
StartUpsDataFrame
```

### Data pr√¶senteret som kurver

```python
{
    StartUpsDataFrame[['Profit','R&D Spend','Administration','Marketing Spend']]
    .sort_values(by='Profit',ascending=False)
    .plot(title="Curves for Start-ups",
          xlabel="Company index sorted on profit",
          ylabel="$",
          cmap='Dark2',
          style=["--",".-","-+","-*"]  )
};
```

###Data pr√¶senteret som s√∏jlediagrammer

```python
{
    StartUpsDataFrame[['Profit','R&D Spend']]
    .sort_values(by='Profit', ascending=False)
    .plot(title="Profit as bars",
          kind="bar",
          color="#68AB33")
};
```
