### Displaying data for one categorical variable

Opgaver:


1.   Giver det mening, at se på andet end variablen 'State', når fokus er kategorivariable?



```python
#Importing the dataset
CSVDataFile="https://raw.githubusercontent.com/jpandersen61/Machine-Learning/refs/heads/main/50_Startups.csv"
StartUpsDataFrame = pd.read_csv(CSVDataFile)
StartUpsDataFrame.head()
```

```python
{
    StartUpsDataFrame['State']
    .value_counts()
    .plot(kind='bar',
          title='Number of Startups by State',
          xlabel='State',
          ylabel='Number of Startups')
};
```

```python
StartUpsDataFrame[['State']].value_counts().plot.pie(y='State', figsize=(5, 5))
```

### Displaying data for two categorical variables

```python
# Sample DataFrame
data = {
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Smoker': ['Yes', 'No', 'Yes', 'No', 'No']
}

df = pd.DataFrame(data)
df
```

```python
pd.crosstab(df['Gender'],df['Smoker'])
```
