I denne sektion skal du lave udtræk af registrerede kunder i e-conomic samt oprette, updatere og slette kunder jf. dokumentationen: https://restdocs.e-conomic.com/#customers

```python
newCustomer = {
  'customerNumber' : customerNumber,
  'currency': 'DKK',
  'customerGroup': defaultCustomerGroup,
  'name': 'Test Customer',
  'paymentTerms': defaultPaymentTerms,
  'vatZone': defaultVatZone
}
```

```python
url = "https://restapi.e-conomic.com/customers"
response = requests.post(url=url, json=newCustomer, headers = headers)
response.reason
```
