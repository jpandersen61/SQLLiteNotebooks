### Hent aktuel vejr for New York

```python
#Konstruer endpoint for aktuelt New York vejr
NewYorkWeatherURL=f"http://api.weatherstack.com/current?access_key={AccessKey}&query=New York"
NewYorkWeatherURL
```

```python
#Brug WeatherStack REST API til at hente aktuelle vejrdata for New York
response = requests.get(NewYorkWeatherURL)
NewYorkWeatherJSON=response.json()
NewYorkWeatherJSON
```

```python
#JSON Object bliver til en dictionary i Python (dict)
type(NewYorkWeatherJSON)
```
